# Matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib as matplot
import numpy as np
print(np.__version__)
print(matplot.__version__)
```

```python
X = np.linspace(0, 10, 15)
X
```

```python
Y = X**2
Y
```

```python
plt.plot(X, Y)
plt.xlabel('X (floating-point values)', fontsize = 15)
plt.ylabel('Y (squares of X)', fontsize = 15)
plt.title('Sample Plot', fontsize = 15)
```

## Subplot
This is used for creating the multiplots on the same canvas.

```python
#Argument takes - (1: Number of rows | 2: Number of columns | 1: The plot number we are refering)
plt.subplot(1, 2, 1)
plt.plot(X, Y, color = 'red')

plt.subplot(1, 2, 2)
plt.plot(Y, X, color = 'black')
```

## Object oriented method

```python
fig = plt.figure()
#getting blank axes (canvas). add_axes takes an arguement of left, bottom, width and height
axes = fig.add_axes([0.1,0.5,0.8,0.8])
#Ploting the values
axes.plot(X, Y)
#set labels
axes.set_xlabel('X (floating-point values)', fontsize = 15)
axes.set_ylabel('Y (squares of X)', fontsize = 15)
axes.set_title('Plot usign Object', fontsize = 15)
```

## Lets try to create canvas and see how their position change with respect to their values

```python
#Both created axes will merge as their values are same
fig = plt.figure()
#
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8], label="axes1")
axes2 = fig.add_axes([0.1, 0.5, 0.8, 0.8], label="axes2")
```

```python
# The first one is 0.1 from left side where as axes2 is 0.2 from the left
# axes1 : 0.5 from the bottom whereas axes2: 0.2 from the bottom.
# Same width and height
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.2, 0.8, 0.8])
```

```python
# The first one is 0.1 from left side where as axes2 is 0.4 from the left
# axes1 : 0.5 from the bottom whereas axes2: 0.6 from the bottom.
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8])
axes2 = fig.add_axes([0.4, 0.6, 0.4, 0.5])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.9, 0.3, 0.2])

axes1.plot(X, Y, color='black')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('Outer_plot')

axes2.plot(X, Y, color='red')
# axes2.set_xlabel('X2')
# axes2.set_ylabel('Y2')
axes2.set_title('Inner_plot')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
#To fix overlapping
plt.tight_layout()
axes
```

```python
# This 'axes' is iterable
fig, axes = plt.subplots(nrows=1, ncols=2)
for i in axes:
    i.plot(X, Y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(X, Y)
axes[1].plot(Y, X)
```

## Figure Size / Aspect ratio / DPI

```python
fig = plt.figure(figsize=(5, 5))
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8])
axes1.plot(X, Y)
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('Outer_plot')
```

```python
fig = plt.figure(figsize=(10, 5))
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8])
axes1.plot(X, Y)
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('Outer_plot')
```

## Legend

```python
fig = plt.figure(figsize=(10,5))
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
axes1.plot(X, X**2,label='x-square')
axes1.plot(X, X**3,label='x-cube')
#loc is used to select location for legend
axes1.legend(loc=0)
```

## Setting linewidth, transparency & linestyle
'linestyle' supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'  

```python
fig = plt.figure(figsize=(10,5))
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8])
axes1.plot(X, Y, color='green', linewidth=5)
```

```python
fig = plt.figure(figsize=(10,5))
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8])
axes1.plot(X, Y, color='green', linewidth=5, alpha=0.5)
```

```python
fig = plt.figure(figsize=(10,5))
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8])
axes1.plot(X, Y, color='green', linewidth=5, linestyle='--')
```

```python
fig = plt.figure(figsize=(10,5))
axes1 = fig.add_axes([0.1, 0.5, 0.8, 0.8])
axes1.plot(X, Y, color='green', linewidth=5, linestyle='dotted')
```

## Marker

```python
#To markout each point
fig = plt.figure(figsize=(10,5))
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
axes1.plot(X, Y,color='green',
                linewidth=2,
                marker='o',
                markersize=10,
                markerfacecolor='red',
                markeredgewidth=3,
                markeredgecolor='yellow')
```

## Limit the plot

```python
fig = plt.figure(figsize=(15, 10))
axes1 = fig.add_axes([0.1,0.5,0.8,0.8], label="axes1")
axes1.plot(X, Y, color='red', linewidth=2)
```

```python
# To get part of the above graph say x from 0 to 2
fig = plt.figure(figsize=(15, 10))
axes2 = fig.add_axes([0.1,0.5,0.8,0.8], label="axes2")
axes2.plot(X, Y, color='red', linewidth=2)
axes2.set_xlim([0,2])
```

```python
fig = plt.figure(figsize=(15, 10))
axes3 = fig.add_axes([0.1,0.5,0.8,0.8], label="axes3")
axes3.plot(X, Y, color='red', linewidth=2)
axes1.set_xlim([0,2])
axes1.set_ylim([0,10])
```
